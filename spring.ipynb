Scraping data from website

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url ='https://stardewvalleywiki.com/Spring'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table', class_ = 'wikitable')
table_Festival = table[0]
table_Birthdays= table [2]
table_Single_Harvest = table[3]
table_Multiple_Harvests = table[4]
table_Forage = table[5]
table_Fish = table[6]
titles = table_Single_Harvest.find_all('th')
```

```python
table_Single_Harvest_titles = [title.text.strip() for title in titles]
print(table_Single_Harvest_titles)
```

```python
df = pd.DataFrame(columns = table_Single_Harvest_titles)

df
```

```python
column_data = table_Single_Harvest.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    table_Single_Harvest_individual_row = [data.text.strip() for data in row_data]
    print(table_Single_Harvest_individual_row)
    length = len(df)
    df.loc[length] = table_Single_Harvest_individual_row
```
